```python
import re
import pandas as pd
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.tokenize import word_tokenize
import nltk

nltk.download("punkt")
```

```python
# Membaca file Excel
file_path = "dataset/review_kaos_shopee.xlsx"
data = pd.read_excel(file_path)

# Menampilkan lima baris pertama untuk memeriksa struktur data
print(data.head())
```

```python
def text_cleaning(text):
    # Inisialisasi StopWordRemover
    factory_stopword = StopWordRemoverFactory()
    stopword = factory_stopword.create_stop_word_remover()

    # Inisialisasi Stemmer
    factory_stemmer = StemmerFactory()
    stemmer = factory_stemmer.create_stemmer()

    # Lowercase teks
    text = text.lower()

    # Hapus karakter khusus, tautan, dan karakter yang tidak relevan
    text = re.sub(r"https?://\S+|www\.\S+", " ", text)
    text = re.sub(r"<.*?>", " ", text)
    text = re.sub(r"[^a-zA-Z\s]", " ", text)

    # Hapus stopwords
    without_stopword = stopword.remove(text)

    # Lakukan stemming
    stemmed_text = stemmer.stem(without_stopword)

    return stemmed_text
```

```python
# Membersihkan kolom 'txt' menggunakan fungsi text_cleaning
data["txt"] = data["txt"].apply(text_cleaning)
```
